## ROC curves for different averaging ranges

(Full documented "paz" module can be found in this same repository.)

```python
import paz
```

```python
df = paz.read_and_merge('paz_metadata_actualitzat.pkl',
                       'paz_profiles_actualitzat.pkl')
df = paz.clean_dataframe(df)
```

```python
#for different sets of initial and final heights, plot the ROC curve
initial = [0.1,0.5]
final = [2.0,5.0]

for hi in initial:
    for hf in final:
        paz.plotROC(df,hi,hf,0.95)
        
```
