```python
his = np.arange(0.1,6.,0.2)
hfs = np.arange(2.,19.,0.2)
matrix = np.zeros([111,211])
results=[]
print(matrix.max())
for hi in tqdm(his):
    for hf in hfs:
        
        i = int(10.*hi)
        j = int(10.*hf)
        
        if hi<hf:
            array = do.getF1score(tropics_df,hi,hf,0.90)
            results.append(array)

            
```

```python
y=np.array([np.array(element) for element in results])
f1scores = y[:,4]
x = np.argmax(f1scores)
print(y[x])
```

```python
his = np.arange(0.1,6.,0.2)
hfs = np.arange(5.,21.,0.2)
matrix = np.zeros([111,211])
results=[]
print(matrix.max())
for hi in tqdm(his):
    for hf in hfs:
        
        i = int(10.*hi)
        j = int(10.*hf)
        
        if hi<hf:
            array = do.getF1score(extratropics_df,hi,hf,0.90)
            matrix[i][j] = array[4]
            results.append(array)

            
```

```python
y=np.array([np.array(element) for element in results])
f1scores = y[:,4]
x = np.argmax(f1scores)
print(y[x])
```
