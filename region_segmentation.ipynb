## At the end of this notebook you can see the performance of the optimal average in tropical vs extratropical regions
(Full documented "paz" module can be found in this same repository.)

```python
import paz
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse as sparse
from matplotlib.font_manager import FontProperties
from tqdm.notebook import tqdm
import plotly.express as px

font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
```

```python
df = paz.read_and_merge('paz_metadata_actualitzat.pkl',
                       'paz_profiles_actualitzat.pkl')
df = paz.clean_dataframe(df)
```

## the next function segmentates the dataset based on latitude values

```python
tropics_df,extratropics_df = paz.separate_tropics(df)
```

```python
his = np.arange(0.1,11.,0.1)
hfs = np.arange(1.,21.,0.1)
percentiles = np.linspace(0.90,0.99,10)
results_tropics=[[],[],[],[],[],[],[],[],[],[]] #all the F1 scores for each percentile will be 
results_extra=[[],[],[],[],[],[],[],[],[],[]] #saved in each of the 10 lists inside results
          
```

```python
results_extra=[[],[],[],[],[],[],[],[],[],[]] #saved in each of the 10 lists inside results
```

### tropics_df

```python
i = 0 
for p in tqdm(percentiles):
    matrix = np.zeros([111,211])
    print(matrix.max())
    for hi in tqdm(his):
        for hf in hfs:

            if hi<hf:
                array = paz.getF1score(tropics_df,hi,hf,p)
                results_tropics[i].append(array)
    i = i+1
```

### extratropics_df

```python
i = 0 
for p in tqdm(percentiles):
    matrix = np.zeros([111,211])
    print(matrix.max())
    for hi in tqdm(his):
        for hf in hfs:

            if hi<hf:
                array = paz.getF1score(extratropics_df,hi,hf,p)
                results_extra[i].append(array)
    i = i+1
```

```python
y=np.array([np.array(element) for element in results_extra[9]]) #aqui tendras que aadir un "for element in results[0]" (para el primer percentile)
f1scores = y[:,4]
x = np.argmax(f1scores)
print(y[x])
```

```python
p1,r1,p,r,truth,label1 = paz.getPrecisionRecall(extratropics_df,0.2,7.7,0.99)
print(p,r)
```

```python
his = np.arange(0.1,6.,0.2)
hfs = np.arange(5.,21.,0.2)
matrix = np.zeros([111,211])
results=[]
print(matrix.max())
for hi in tqdm(his):
    for hf in hfs:
        
        i = int(10.*hi)
        j = int(10.*hf)
        
        if hi<hf:
            array = paz.getF1score(extratropics_df,hi,hf,0.90)
            matrix[i][j] = array[4]
            results.append(array)

            
```

```python
y=np.array([np.array(element) for element in results])
f1scores = y[:,4]
x = np.argmax(f1scores)
print(y[x])
```

```python
p1,r1,p11,r11,truth,label1 = paz.getPrecisionRecall(tropics_df,3.6,18.2,0.94)
no_skill = len(truth[truth==1]) / len(truth)

p2,r2,p22,r22,truth2,label2 = paz.getPrecisionRecall(extratropics_df,1.0,7.6,0.94)

plt.figure(figsize=(5,5), tight_layout=True)

plt.scatter(p1,r1,label='Tropical regions',s=0.5,c='crimson')

plt.scatter(p2,r2,label='Extratropical regions',s=0.5,c='steelblue')

plt.scatter(p11,r11,s=40,c='black',label='Best F1 Score')



plt.plot([0, 1], [no_skill, no_skill], linestyle='--', c='black',label='no skill')



plt.xlabel('Recall', 
           fontproperties=font,
           fontsize=23)
plt.ylabel('Precission',
           fontproperties=font,
           fontsize=23)

plt.xticks(ticks=[0,0.25,0.5,0.75,1], 
           labels=['0','0.25','0.5','0.75','1.0'],
           fontproperties=font, 
           fontsize=20)
plt.yticks(ticks=[0,0.25,0.5,0.75,1], 
           labels=['0','0.25','0.5','0.75','1.0'],
           fontproperties=font, 
           fontsize=20)
plt.grid()
```
