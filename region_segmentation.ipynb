```python
import do_functions as do
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse as sparse
from matplotlib.font_manager import FontProperties
from tqdm.notebook import tqdm
import plotly.express as px

font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
```

```python
df = do.read_and_merge('paz_metadata_actualitzat.pkl',
                       'paz_profiles_actualitzat.pkl')
df = do.clean_dataframe(df)
```

```python
tropics_df,extratropics_df = do.separate_tropics(df)
```

```python
his = np.arange(0.1,11.,0.1)
hfs = np.arange(1.,21.,0.1)
percentiles = np.linspace(0.90,0.99,10)
results_tropics=[[],[],[],[],[],[],[],[],[],[]] #all the F1 scores for each percentile will be 
results_extra=[[],[],[],[],[],[],[],[],[],[]] #saved in each of the 10 lists inside results
          
```

```python
results_extra=[[],[],[],[],[],[],[],[],[],[]] #saved in each of the 10 lists inside results
```

### tropics_df

```python
i = 0 
for p in tqdm(percentiles):
    matrix = np.zeros([111,211])
    print(matrix.max())
    for hi in tqdm(his):
        for hf in hfs:

            if hi<hf:
                array = do.getF1score(tropics_df,hi,hf,p)
                results_tropics[i].append(array)
    i = i+1
```

### extratropics_df

```python
i = 0 
for p in tqdm(percentiles):
    matrix = np.zeros([111,211])
    print(matrix.max())
    for hi in tqdm(his):
        for hf in hfs:

            if hi<hf:
                array = do.getF1score(extratropics_df,hi,hf,p)
                results_extra[i].append(array)
    i = i+1
```

```python
y=np.array([np.array(element) for element in results_extra[9]]) #aqui tendras que aadir un "for element in results[0]" (para el primer percentile)
f1scores = y[:,4]
x = np.argmax(f1scores)
print(y[x])
```
